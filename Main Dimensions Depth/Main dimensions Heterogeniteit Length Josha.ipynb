Heterogeniteit testen op Length

```python
from scipy.stats import ttest_ind
import pandas as pd

import os
os.chdir('j:\Studio De Voogt\Specialists & NA\Feadship Database\Feadship Python\JOSHA_workingstudent_2024\Main Dimensions Depth')
current_path = os.getcwd()
print(current_path)

# Load dataset
tabel = pd.read_excel('Main Dimensions Cleaned.xlsx')
tabel = tabel.dropna(subset=['Beam Moulded', 'Depth'])  # Remove missing values

# Range of values for Length
highest = tabel['Beam Moulded'].max()
print(highest) # 22.5
lowest = tabel['Beam Moulded'].min()
print(lowest) # 2
```

Testen van optimale treshold voor Length

```python
# Definieer het bereik voor mogelijke grenswaarden
min_threshold = 2
max_threshold = 22.5
step_size = 1

# Variabelen om de beste grens en p-waarde op te slaan
best_threshold = None
lowest_p_value = 1  # Start met een p-waarde van 1 (hoogst mogelijke p-waarde)

# Itereer door elke mogelijke grenswaarde
for threshold in range(min_threshold, max_threshold + 1, step_size):
    # Maak de groepen op basis van de huidige grenswaarde
    group_above = tabel[tabel['Beam Moulded'] > threshold]['Depth']
    group_below = tabel[tabel['Beam Moulded'] <= threshold]['Depth']
    
    # Controleer of er voldoende waarden in beide groepen zijn
    if len(group_above) > 1 and len(group_below) > 1:
        # Voer een t-test uit
        t_stat, p_value = ttest_ind(group_above, group_below, equal_var=False)  # Welch's t-test
        
        # Update de beste grenswaarde indien deze p-waarde lager is
        if p_value < lowest_p_value:
            lowest_p_value = p_value
            best_threshold = threshold

# Print de optimale grenswaarde en de bijbehorende p-waarde
print(f"Beste grenswaarde: {best_threshold} meter")
print(f"Bijbehorende p-waarde: {lowest_p_value}")

# Interpretatie van de beste grenswaarde
if lowest_p_value < 0.05:
    print(f"De grenswaarde van {best_threshold} meter geeft een significant verschil in diepte.")
else:
    print(f"Er is geen significant verschil gevonden binnen het bereik 45-70 meter.")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Voeg een kolom toe aan de dataset die aangeeft of een schip boven of onder de beste grenswaarde ligt
tabel['Length_Group'] = ['Above 52m' if length > 52 else 'Below or Equal 52m' for length in tabel['Length Over All']]

# Maak een boxplot om de diepteverdeling in beide groepen te vergelijken
plt.figure(figsize=(10, 6))
sns.boxplot(x='Length_Group', y='Depth', data=tabel)
plt.title('Vergelijking van diepte tussen schepen boven en onder 52 meter lengte')
plt.xlabel('Lengte van het schip')
plt.ylabel('Diepte')
plt.show()

# Optioneel: Violin plot voor nog meer details over de verdeling
plt.figure(figsize=(10, 6))
sns.violinplot(x='Length_Group', y='Depth', data=tabel)
plt.title('Diepteverdeling tussen schepen boven en onder 52 meter lengte')
plt.xlabel('Lengte van het schip')
plt.ylabel('Diepte')
plt.show()
```
